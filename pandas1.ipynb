### 시리즈(series)
데이터가 순차적으로 1차원 배열(리스트)이다.<br>
인덱스(index)는 데이터값(value)와 일대일 대응함

```python
# pandas (외부 라이브러리) 설치
pip install pandas
```

```python
import pandas

s = pandas.Series(["사과", "감", "바나나"])
print(s)

print(s.index)
print(type(s))

# 인덱스 1, 2, 3으로 변경
s1 = pandas.Series(["사과", "감", "바나나"], index=[1, 2, 3])
print(s1)

import pandas as pd
s = pd.Series([3000, 2000, 1000])
s.index = [1, 2, 3]
print(s)
```

### DataFrame
series를 결합한 형태이다.<br>
행과 열을 가진 2차원 리스트 형태이다.

```python
import pandas as pd

name = pandas.Series(["사과", "감", "바나나"])
price = pd.Series([3000, 2000, 1000])

data = {"name": name, "price": price}

# 데이터 프레임 만기기
df = pd.DataFrame(data)
print(df)
type(df)

df.columns
```

```python
# 방법2 - 데이터프레임 만들기
data = {"name": ["사과", "감", "바나나"], "price": [3000, 2000, 1000]}
df = pd.DataFrame(data)
print(df)

# 칼럼 이름 변경 - rename()
df_new = df.rename(columns={"name": "이름", "price": "가격"})
print(df_new)

# 칼럼 추가
df_new["중량"] = "100g"
print(df_new)

# 데이터의 정보
df_new.info()

# to_csv("파일이름")   df_new를 csv 파일로 저장하기(변환)
df_new.to_csv("data/fruit.csv", index=False)   # 추가 인덱스를 없게



# read_csv("파일이름") : csv 파일 읽기
fruit = pd.read_csv("fruit.csv")
print(fruit)
```
