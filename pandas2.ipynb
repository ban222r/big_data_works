```python
# 결측치 대체(수치형)  - 평균값(mean()), 중앙값(median())
# 결측치 대체(범주형, 문자형)  - 최민값(mode())

import pandas as pd
df = pd.DataFrame({"A" : [1, None, 3, 5, 4], "B" : [4, 5, None, 7, 8], "C" : [7, None, 9, 10, None], "D" : ["sky", "river", "sky", None, "sky"]})
print(df)

# sum_val = df['C'].sum()
# print(sum_val)

# mean_val = df['C'].mean()
# print(mean_val)

# 대체하는 함수  - fillna()
# df['C'] = df['C'].fillna(round(mean_val, 1))   # df['c']로 받는 것이 중요
# print(df)

# 최빈값  - mode()
most_freq = df['D'].mode()[0]      # 마지막 index를 넣어주어야 함
print(most_freq)
df['D'] = df['D'].fillna(most_freq)
print(df)
```

```python
# 범주형 데이터 만들기
df_car = pd.DataFrame({"car" : ['Sedan', 'SUV', 'Sedan', 'SUV', 'SUV', 'SUV', 'Sedan', 'Sedan', 'Sedan', 'Sedan', 'Sedan'],
                       "size" : ['S', 'M', 'S', 'S', 'M', 'M', 'L', 'S', 'S', 'M', 'S']})
print(df_car)

# 고유값(도메인) 개수  - nunique(), unique()
df_car.nunique()

print(df_car['car'].unique())
print(df_car['size'].unique())

# 도메인(범주)의 개수  - value_counts()
print(df_car['car'].value_counts())
print(df_car['size'].value_counts())

# 요약 통계
df_car.describe(include="O")   # O(대문자) = object
```

```python
# 오름 차순, 내림 차순 정렬   - sort_values()
# ascending=True (오름차순) ,  ascending=False(내림차순)

df_cafe = pd.read_csv("./data/cafe2.csv")
df_cafe

# 칼럼 1개 기준 정렬
# df_cafe.sort_values("가격")    # 오름차순 디폴트
df_cafe.sort_values("가격", ascending=False)   # 내림차순

# 칼럼이 2개 이상의 기준 정렬
df_cafe.sort_values(["가격", "메뉴"], ascending=[False, True])    # 가격은 내림차순, 메뉴는 오름차순

# 인덱스 기준 정렬  - sort_index()
df_cafe.sort_index()    # 오름차순 디퐅트
df_cafe.sort_index(ascending=False)  # 내림차순

# 특정 칼럼으로 인덱스 설정  - set_index()        # 인덱스 초기화 - reset_index(drop=True)
df_cafe.set_index("메뉴")
```
